# spb

```python
import pandas as pd
import io
import hashlib
from google.colab import files
from IPython.display import display

# Meminta user untuk mengunggah file xls dari inaportnet
print("Silakan unggah file SPB (ext xls) yg diunduh dari Inaportnet:")
uploaded = files.upload()

all_dfs = []

for filename in uploaded.keys():
    try:
        file_bytes = uploaded[filename]
        file_hash = hashlib.sha256(file_bytes).hexdigest()
        print(f"\nReading uploaded file: {filename}")

        dfs = pd.read_html(io.BytesIO(file_bytes))
        if dfs:
            df = dfs[0]
            # Menambahkan kolom source_file
            df['source_file'] = filename
            print(f"Berhasil mengimpor: {filename}")
            all_dfs.append(df)
    except Exception as e:
        try:
            df = pd.read_excel(io.BytesIO(uploaded[filename]), engine='xlrd')
            df['source_file'] = filename
            all_dfs.append(df)
            print(f"Berhasil mengimpor (Excel mode): {filename}")
        except Exception as e2:
            print(f"Gagal membaca {filename}: {e2}")

if all_dfs:
    combined_spb = pd.concat(all_dfs, ignore_index=True)
    print("\nSemua file berhasil digabungkan dengan kolom 'source_file'.")
    display(combined_spb.head(5))
else:
    print("Tidak ada data yang berhasil diimpor.")
```

```python
print(f"Jumlah baris sebelum penghapusan duplikat: {len(combined_spb)}")

duplicate_rows = combined_spb.duplicated().sum()

if duplicate_rows > 0:
    print(f"Ditemukan {duplicate_rows} baris duplikat di combined_spb. Membersihkan...")
    combined_spb = combined_spb.drop_duplicates(keep='first')
    print(f"Jumlah baris setelah penghapusan duplikat: {len(combined_spb)}")
    print("Duplikat berhasil dihapus.")
else:
    print("Tidak ditemukan baris duplikat di combined_spb.")

display(combined_spb.head())
```

```python
display(combined_spb.tail(5))
```

```python
import pandas as pd

# 1. Merubah kolom 'No' menjadi string
combined_spb['No'] = combined_spb['No'].astype(str)

# 2. Merubah kolom tanggal menjadi datetime
# Format standar pandas to_datetime biasanya otomatis mendeteksi format yang ada
# errors='coerce' digunakan agar data yang tidak valid berubah menjadi NaT (Not a Time)
date_columns = ['Tanggal Tiba', 'Tanggal Terbit', 'Tanggal Tolak']

for col in date_columns:
    combined_spb[col] = pd.to_datetime(combined_spb[col], errors='coerce')

# Verifikasi perubahan tipe data
print("Tipe data kolom setelah perubahan:")
print(combined_spb[['No', 'Tanggal Tiba', 'Tanggal Terbit', 'Tanggal Tolak']].dtypes)
display(combined_spb.head())
```

```python
from google.colab import files
import pandas as pd

def export_dataframe_by_name():
    """
    Meminta input nama variabel dataframe dari user,
    lalu mengekspornya ke Excel.
    """
    df_name = input("Masukkan nama variabel dataframe yang ingin diekspor (contoh: combined_spbspog_dermaga): ").strip()

    # Memeriksa apakah variabel ada di namespace global
    if df_name in globals():
        df = globals()[df_name]

        if isinstance(df, pd.DataFrame):
            filename = f"{df_name}.xlsx"
            df.to_excel(filename, index=False)
            files.download(filename)
            print(f"Berhasil! Dataframe '{df_name}' sedang diunduh sebagai '{filename}'.")
        else:
            print(f"Error: '{df_name}' bukan merupakan Pandas DataFrame.")
    else:
        print(f"Error: Variabel '{df_name}' tidak ditemukan.")

# Jalankan fungsi
export_dataframe_by_name()
```

```python
import pandas as pd

# Memisahkan dataframe berdasarkan kolom 'SPK'
# 1. Baris yang memiliki SPK (tidak NaN)
combined_spbspog_dermaga_SPK = combined_spbspog_dermaga[combined_spbspog_dermaga['SPK'].notna()].copy()

# 2. Baris yang tidak memiliki SPK (NaN)
combined_spbspog_dermaga_NoSPK = combined_spbspog_dermaga[combined_spbspog_dermaga['SPK'].isna()].copy()

# Verifikasi jumlah baris
print(f"Total baris di combined_spbspog_dermaga: {len(combined_spbspog_dermaga)}")
print(f"Baris dengan SPK: {len(combined_spbspog_dermaga_SPK)}")
print(f"Baris tanpa SPK: {len(combined_spbspog_dermaga_NoSPK)}")

print("\nPreview Data dengan SPK:")
display(combined_spbspog_dermaga_SPK.head(3))

print("\nPreview Data tanpa SPK:")
display(combined_spbspog_dermaga_NoSPK.head(3))
```

```python
import numpy as np

# Definisikan kondisi untuk masing-masing kriteria
cond_pandu_tunda = (combined_spbspog_dermaga_NoSPK['GT'] >= 500) & (combined_spbspog_dermaga_NoSPK['LOA'] >= 70)
cond_pandu_only = (combined_spbspog_dermaga_NoSPK['GT'] >= 500) & (combined_spbspog_dermaga_NoSPK['LOA'] < 70)

# Mengisi kolom RedFlag1 berdasarkan hierarki kondisi
combined_spbspog_dermaga_NoSPK['RedFlag1'] = ""
combined_spbspog_dermaga_NoSPK.loc[cond_pandu_only, 'RedFlag1'] = "Seharusnya Pandu karena GT>=500"
combined_spbspog_dermaga_NoSPK.loc[cond_pandu_tunda, 'RedFlag1'] = "Seharusnya Pandu&Tunda karena GT>=500 & LoA >=70m"

# Filter untuk melihat baris yang terkena RedFlag1
redflag_results = combined_spbspog_dermaga_NoSPK[combined_spbspog_dermaga_NoSPK['RedFlag1'] != ""]

print(f"Jumlah baris yang masuk kriteria RedFlag1: {len(redflag_results)}")
display(redflag_results[['No Layanan', 'Kapal', 'GT', 'LOA', 'STATUS?', 'RedFlag1']].head(10))
```
